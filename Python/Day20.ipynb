```python
import pandas as pd
```

```python
bc_data = pd.read_csv('data.csv')
```

```python
bc_data.head()
```

```python
# Drop unnecessary columns
bc_data.drop('id', axis=1, inplace=True)
```

```python
# Encode the target variable 'diagnosis' (M or B)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
bc_data['diagnosis'] = le.fit_transform(bc_data['diagnosis'])
```

```python
bc_data.head()
```

```python
print(bc_data.info())
```

```python
print(bc_data['diagnosis'].value_counts())
```

```python
# 1 Malignant
# 0 Benign 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Data Visualization
sns.pairplot(bc_data, hue='diagnosis', vars=['radius_mean', 'texture_mean', 'area_mean'])
plt.show()
```

```python
# Example
sns.countplot(x='diagnosis', data=bc_data)
plt.title('Distribution of Diagnosis')
plt.show()
```

```python
# Correlation
correlation_matrix = bc_data.corr()
plt.figure(figsize=(23, 19))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation')
plt.show()
```

```python
plt.figure(figsize=(24, 12))
for i, column in enumerate(bc_data.select_dtypes(include='number').columns, 1):
    plt.subplot(2, 4, i)
    sns.boxplot(x='diagnosis', y=column, data=bc_data)
    plt.title(f'Boxplot of {column} by Diagnosis')
plt.tight_layout()
plt.show()
```
