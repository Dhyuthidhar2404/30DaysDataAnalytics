```python
#build dataframe
df = pd.DataFrame(columns=['video_id','video_title','upload_date','view_count','like_count','favoriteCount','comm_count'])
```

```python
df
```

```python
#parsing through the data
for video in response['items']:
    if video['id']['kind'] == "youtube#video":
        video_id = video['id']['videoId']
        title_video = video['snippet']['title']
        video_title = str(title_video).replace("&","")
        upload_date = video['snippet']['publishedAt']
        upload_date = str(upload_date).split("T")[0]
        
        url_stat = "https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&part=statistics&key="+API_key
        response_video_stat = requests.get(url_stat).json()
        
        view_count = response_video_stat['items'][0]['statistics']['viewCount']
        like_count = response_video_stat['items'][0]['statistics']['likeCount']
        favoriteCount = response_video_stat['items'][0]['statistics']['favoriteCount']
        comm_count = response_video_stat['items'][0]['statistics']['commentCount']
        
        #save the data into df
        df = df.append({'video_id':video_id,'video_title':video_title,
                         "upload_date":upload_date,"view_count":view_count,
                         "like_count":like_count,"favoriteCount":favoriteCount,
                         "comm_count":comm_count},ignore_index=True)
```
